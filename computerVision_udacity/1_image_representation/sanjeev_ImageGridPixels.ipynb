```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
```

```python
image = mpimg.imread('images/waymo_car.jpg')

print("Image dimensions", image.shape)
```

```python
#Grayscale Image

gray_scale = cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
plt.imshow(gray_scale,cmap='gray')
```

```python
#Printing Specific GrayScale

x = 400
y = 300

pixel = gray_scale[y,x]
print(pixel)
```

```python
max_value = np.amax(gray_scale)
min_value = np.amin(gray_scale)
```

```python
print("MaX Pixel Value is: ", max_value)
print("MiN Pixel Value is: ", min_value)
```
