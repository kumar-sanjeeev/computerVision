## Important points about canny Edge Detector
**ALGO**
* 1. Filter out noises using Gaussian Filter(for imnage smoothing) (Low pass filter)
* 2. Apply the both sobel_x and sobel_y Filter (High pass Filter) to find out strength and direction of the edges
* 3. Applies Non-Maximum Suppression
* 4. Finally uses the hystersis to isolate best edges.

**Hystersis**:
* It is double thresholding Process



### Implementing Canny Edge Detection

```python
import cv2
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

image = mpimg.imread('images/sunflower.jpg')
copy_image = np.copy(image)

# convert into the gray scale
gray = cv2.cvtColor(copy_image,cv2.COLOR_RGB2GRAY)
plt.title("Gray Scaled Image")
plt.imshow(gray, cmap='gray')
```

```python
# Define the high and low threshold for hystersis
# Always choose the lower and higer threshold in the ratio of 1:2 to 1:3 for better results

lower = 120
higher = 240

edges = cv2.Canny(gray,lower,higher)
plt.imshow(edges, cmap='gray')
plt.xticks([])
plt.yticks([])
```
