## Low Pass FIlter---> For removing noise/smoothing the images
### For Ex: Gaussain Blur

## Hight Pass Filter-->For Edge Detection
### For Ex: Sobel Filter



Average Filtering:
* Blur/smooth image
* Block high frequency parts of image

Gaussian Filtering:
* Blur/smooth image
* Block high frequency parts of image
* Preserve Edges

```python
import cv2
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

```python
image = mpimg.imread('images/brain_MR.jpg')
copy_image = np.copy(image)

grayimage = cv2.cvtColor(copy_image, cv2.COLOR_RGB2GRAY)
plt.imshow(grayimage, cmap='gray')
```

```python
# Apply gausssian Blur

gray_blur = cv2.GaussianBlur(grayimage,(5,5),0)

f,(ax1,ax2) = plt.subplots(1,2,figsize =(20,10))
ax1.set_title("Original Image")
ax1.imshow(grayimage, cmap='gray')

ax2.set_title("Blurred Image")
ax2.imshow(gray_blur,cmap='gray')
```

Test the performance with High Pass Filter like sobel filter

```python
# High Pass Filter

sobel_x = np.array([[-2,0,2],
                   [-1,0,1],
                   [-2,0,2]])

without_blur = cv2.filter2D(grayimage, -1, sobel_x)
with_blur = cv2.filter2D(gray_blur, -1, sobel_x)

f,(ax1,ax2)= plt.subplots(1,2,figsize =(20,10))
ax1.set_title("Without Blur")
ax1.imshow(without_blur,cmap='gray')

ax2.set_title("With Blur")
ax2.imshow(with_blur, cmap='gray')
```
