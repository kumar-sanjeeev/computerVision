```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import cv2

image = mpimg.imread('images/phone.jpg')
copy_image = np.copy(image)
plt.imshow(image)
```

Perform the Edge Detection

```python
# convert the image into grayscale before putting the canny edge detector

gray = cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
plt.imshow(gray, cmap='gray')
```

```python
# Edge Detection using canny Edge Detector
low_threshold = 50
high_threshold = 100
edge = cv2.Canny(gray, low_threshold, high_threshold)
plt.imshow(edge, cmap='gray')
```

Hough line Detection

```python
# Step 1: Find lines using Hough Transform
rho = 1
# rho and theta are hough variable that defines the resolution of detection, we setted this as 1 pixel and 1 degree
theta = np.pi/180
threshold = 60 # min no of hough intersection to define the line
min_line_length = 100
max_line_gap = 5

# it will return all the detected hough lines
# each line is an array of Four points x1,y1,x2,y2.
lines = cv2.HoughLinesP(edge, rho, theta, threshold, np.array([]), min_line_length, max_line_gap)


# iterate over each output lines and draw on the copy image

for line in lines:
    for x1,y1,x2,y2 in line:
        cv2.line(copy_image,(x1,y1),(x2,y2),(255,0,0), 5)

plt.imshow(copy_image)
```
