# Learning How to create a filter and what is the use case of Kernel in the Filter

```python
import cv2
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('images/city_hall.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
    # Make a copy of image and convert it into the gray scale
    copy_image = np.copy(image)
    gray_image = cv2.cvtColor(copy_image, cv2.COLOR_RGB2GRAY)
    plt.imshow(gray_image, cmap='gray')
```

Creating a Custom Kernel

```python
s_kernel = np.array([[-1,0,1],
                    [-2,0,2],
                    [-1,0,1]])
```

```python
#Applying the custom filter to the image
# If want to detect the vertical edges, apply sobel_x filer

filtered_image = cv2.filter2D(gray_image,-1,s_kernel)
plt.imshow(filtered_image, cmap='gray')
```

Create the Binary Image

```python
# create threshold that sets all filetered  pixels to white
# Above a certain threshold

retval, binary_image = cv2.threshold(filtered_image, 100, 255, cv2.THRESH_BINARY)
plt.imshow(binary_image, cmap='gray')
```
